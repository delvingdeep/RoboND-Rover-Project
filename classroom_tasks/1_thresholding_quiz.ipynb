```python
# Import packages
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
```

```python
# define filename and plot the image
filename = 'media/sample.jpg'
image = mpimg.imread(filename)
plt.imshow(image)
plt.show()
```

```python
# inspect image array
print("Image type: {} \nImage shape: {}\nImage min value: {} \nImage max value: {}".format(image.dtype, image.shape, np.min(image), np.max(image)))
```

```python
# isolate each channels
red_channel = np.copy(image)
red_channel[:,:,[1,2]] = 0    # zero out the green and blue channels

green_channel = np.copy(image)
green_channel[:,:,[0,2]] = 0    # zero out the red and blue channels

blue_channel = np.copy(image)
blue_channel[:,:,[0,1]] = 0    # zero out the red and green channels

# plot all the channle images
fig = plt.figure(figsize=(12,3))    # create a figure for plotting

plt.subplot(131)                    # initialize subplot number 1
plt.title('Red')
plt.imshow(red_channel)

plt.subplot(132)                    # initialize subplot number 2
plt.title('Green')
plt.imshow(green_channel)

plt.subplot(133)                    # initialize subplot number 3
plt.title('Blue')
plt.imshow(blue_channel)

plt.show()
```

```python
# color threshold function

def color_thresh(img, rgb_thresh=(0, 0, 0)):
    thresh_img = np.zeros_like(img[:,:,0])    # creates empty array of same size as img with single channel
    thresh_range = (img[:,:,0] > rgb_thresh[0]) & (img[:,:,1] > rgb_thresh[1]) & (img[:,:,2] > rgb_thresh[2])
    thresh_img[thresh_range] = 1
    return thresh_img
```

```python
# color threshold

red_threshold = 160
green_threshold = 160
blue_threshold = 160

rgb_threshold = (red_threshold, green_threshold, blue_threshold)

threshold_image = color_thresh(image, rgb_thresh=rgb_threshold)
plt.imshow(threshold_image, cmap='gray')
plt.title('Threshold image')
plt.show()
```

```python
# save output as image
mpimg.imsave('media/threshold_image', threshold_image, cmap='gray')
```
